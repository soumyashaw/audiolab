```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import load_model
model = r'C:\Users\Dell\Documents\GitHub\real-world-fake-audio\Deep4SNet\Dataset\audio\model_Deep4SNet.h5'
weights_model = r'C:\Users\Dell\Documents\GitHub\real-world-fake-audio\Deep4SNet\Dataset\audio\weights_Deep4SNet.h5'
model = load_model(model)
model.load_weights(weights_model)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
####   Testing from Original Audio Histogram   ####

pathReal = r'C:\Users\Dell\Desktop\audio\Original_audios_hist'
entries = os.listdir(pathReal)
classifiedReal1 = 0
classifiedFake1 = 0

for fileName in entries:
    # Read image file into image variable
    img = cv2.imread(r'C:\Users\Dell\Desktop\audio\Original_audios_hist\\' + fileName)
    img = cv2.resize(img,(150,150)) 
    img = np.reshape(img,[1,150,150,3])  
    classes = model.predict(img, batch_size=19)

    ### DEBUG ###
    print(classes[0])

    if classes[0]>0.5:
      classifiedReal1 += 1
      ### DEBUG ###
      #print(" The voice recording " + fileName +  " is original")
      
    else:
      classifiedFake1 += 1
      ### DEBUG ###
      #print(" The voice recording" + fileName + " is fake")

print("No of Images Classified as Real (Should be Real):", classifiedReal1)
print("No of Images Classified as Fake (Should be Real):", classifiedFake1)


classifiedReal2 = 0
classifiedFake2 = 0

pathFake = r'C:\Users\Dell\Desktop\audio\Fake_audios_hist'
entries = os.listdir(pathFake)

for fileName in entries:
  
  # Read image file into image variable
  img = cv2.imread(r'C:\Users\Dell\Desktop\audio\Fake_audios_hist\\' + fileName)
  img = cv2.resize(img,(150,150)) 
  img = np.reshape(img,[1,150,150,3])  
  classes = model.predict(img, batch_size=19)

  ### DEBUG ###
  print(classes[0])

  if classes[0]>0.5:
    classifiedReal2 += 1
    ### DEBUG ###
    #print(" The voice recording " + fileName +  " is original")
    
  else:
    classifiedFake2 += 1
    ### DEBUG ###
    #print(" The voice recording" + fileName + " is fake")


print("No of Images Classified as Real  (Should be Fake):", classifiedReal2)
print("No of Images Classified as Fake  (Should be Fake):", classifiedFake2)

accuracy = (classifiedReal1 + classifiedFake2)/(classifiedReal1 + classifiedReal2 + classifiedFake1 + classifiedFake2)
print("Accuracy:", accuracy)
```
